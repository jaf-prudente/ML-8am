```python
#-----------------------------------------------------------------------------
#----------------------------------DOS----------------------------------------
# Ahora vamos a escribir una función que reciba una cadena y que regrese en una lista los números en esa cadena.

def problema2a(lista_de_chars):
    
    # Partimos la cadena que recibimos y la guardamos en el arreglo 'lista'. El criterio para partirla es en espacios, podríamos cambiarlo por comas.
    lista = lista_de_chars.split(' ')
    
    # En 'numeritos' guardamos la lista que nos piden. Serán los enteros en el arreglo 'lista' que son completamente numéricos.
    numeritos = [ int(i) for i in lista if i.isnumeric() == True ]
    
    # La función regresa el arreglo 'numeritos'.
    return numeritos

#-----------------------------------------------------------------------------
# En esta parte escribiremos una función que reciba una oración y regrese un diccionario en donde las llaves serán las palabras y los valores serán la longitud de esas palabras.

def problema2b(lista_de_chars):
    
    # Al igual que antes en el arreglo 'lista' guardamos la oración que recibimos pero chopeadita por espacios.
    lista = lista_de_chars.split(' ')
    
    # En 'diccionario' guardamos lo que nos piden, un diccionario donde las llaves son cada palabra de la oración y su corresponciente valor es la longitud de esa palabra.
    diccionario = { i : len(i) for i in lista }
    
    # La función regresa el diccionario que queremos.
    return diccionario
```

```python
#-----------------------------------------------------------------------------
#----------------------------------TRES---------------------------------------
# En esta celda desarrollaremos el tercer ejercicio: crearemos una función que reciba el archivo de temperaturas (en forma de su ruta) y regrese en un diccionario valores estadísticos del mismo.

def problema3(ruta):
    
    # En 'archivo' se guardará literalmente el archivo abierto en la ruta dada. Solo tiene permisos de lectura por si las dudas.
    archivo = open(ruta, 'r')
    
    # En 'datos' guardamos literal el contenido del archivo.
    datos = archivo.readlines()
    
    # El contenido original son carácteres, entonces los convertimos en flotantes y los guardamos en 'temperaturas'.
    temperaturas = [ float(i) for i in datos ]
    
    # Siempre debemos cerrar las puertas que abrimos, así como los archivos que invocamos. 
    archivo.close()
    
    # Aquí usamos directamente funciones de 'numpy' para calcular los valores que nos piden.
    temp_max = np.max(temperaturas)
    temp_min = np.min(temperaturas)
    temp_prm = np.mean(temperaturas)
    temp_med = np.median(temperaturas)
    
    # En 'diccionario' guardamos el conjunto que nos piden.
    diccionario = {'máximo': temp_max, 'mínimo': temp_min, 'promedio': temp_prm, 'mediana': temp_med}
    
    # La función regresa el diccionario deseado.
    return diccionario

# La ruta del archivo que queremos modificar es la siguiente (suponiendo que setá en la misma carpeta que esta notebook). Solo tendríamos que descomentar y llamar a la función para tener lo que queremos.
# ruta = './temperaturas.txt'
```

```python
#-----------------------------------------------------------------------------
#----------------------------------CUATRO-------------------------------------
# El siguiente ejercicio se trata de implementar una función que reciba un arreglo de números y que regrese la desviación estándar de la misma.

def problema4(lista):
    
    # En 'mu' guardamos el promedio de los valores.
    mu = np.mean(lista)
    
    # En 'N'  guardamos la longitud de la muestra (cuántos valores tiene).
    N = len(lista)
    
    # Iniciamos la variable 'suma' en cero, guardará la suma de la diferencia al cuadrado.
    suma = 0
    
    # En este ciclo llenamos justamente la suma que necesitamos
    for i in np.arange(0, len(lista)):
        suma += (lista[i] - mu)**2
    
    # Calculamos directamente la desviación.
    desviacion = np.sqrt(suma/N)
    
    # Esta función regresa la desviacion estándar.
    return desviacion
```

```python
#-----------------------------------------------------------------------------
#----------------------------------SIETE--------------------------------------
# Vamos ahora a escribir una función que simule el lanzamiento de un dado para calcular la probabilidad frecuencial. Recibirá cuántas veces queremos lanzar el dado y cuál es el valor que nos interesa para calcular la probabilidad.
def problema7a(num_de_tiros, valor_favorable):
    
    # En 'casos_favorables' guardaremos cuántas veces ha caído el dado en el valor deseado.
    casos_favorables = 0
    
    # En 'casos_totales' guardaremos cuántas tiradas hemos hecho.
    casos_totales = 0
    
    # Este ciclo es para hacer tantas tiradas como queramos.
    for i in range(num_de_tiros):
        
        # En 'tirada' guardaremos el valor en el que ha caído el dado.
        tirada = rd.choice(range(1, 7))
        
        # Si el valor en el que ha caído el dado es igual al caso favorable entonces aumentamos en uno los casos favorables.
        if tirada == valor_favorable:
            casos_favorables += 1
            
        # En cada corrida del ciclo tiramos una vez el dado, entonces tenemos que aumentar en uno los casos totales.
        casos_totales += 1
    
    # La probabilidad frecuencial se define como el cociente entre los casos favorables y los totales.
    proba = casos_favorables/casos_totales
    
    # Esta función regresa la probabilidad frecuencial.
    return proba

#-----------------------------------------------------------------------------
# Ahora vamos a definir una función que escriba el archivo con datos obtenidos a partir de n simulaciones.
def problema7b(num_de_tiros):
    
    # Definimos una ruta donde se guardará el archivo generado.
    ruta = './problema7b.cvs'
    
    # En la variable 'archivo' gaurdamos justamente el archivo en la ruta que definimos antes con permisos de lectura.
    archivo = open(ruta, 'w')
    
    # Esta es mera información que aparece en la primer línea del archivo. Tiene el simbolo de gato por si se quiere graficar en 'gnuplot'.
    cabezal = '#intento' + '\t' + 'result' + '\t' + 'proba1' + '\t' + 'proba2' + '\t' + 'proba3' + '\t' + 'proba4' + '\t' + 'proba5' + '\t' + 'proba6' + '\n'
    
    # Escribimos esta primer línea con información en el archivo.
    archivo.write(cabezal)
    
    # En la variable 'favorablesi' guardaremos cuántas veces el dado ha caído 'i'.
    favorables1 = 0
    favorables2 = 0
    favorables3 = 0
    favorables4 = 0
    favorables5 = 0
    favorables6 = 0
    
    # En 'totales' guardaremos los tiros totales que hemos dado.
    totales = 0
    
    # Este ciclo se encarga justamente de hacer los 'n' tiros y capturar la información pertinente.
    for i in range(num_de_tiros):
        
        # La variable 'tirada' guarda lo que caiga el dado.
        tirada = rd.choice(range(1, 7))
        
        # En este conjunto de 'if' y 'elif' vamos contando cuántas veces ha caído en los posibles valores. 
        if tirada == 1:
            favorables1 += 1
            
        elif tirada == 2:
            favorables2 += 1
            
        elif tirada == 3:
            favorables3 += 1
        
        elif tirada == 4:
            favorables4 += 1
            
        elif tirada == 5:
            favorables5 += 1
            
        elif tirada == 6:
            favorables6 += 1
        
        # Como dijimos antes aquí guardamos cuántas tiradas hemos dado.
        totales += 1
        
        # Aquí calculamos la probabilidad frecuencial en cada corrida de la iteración.
        proba1 = favorables1/totales
        proba2 = favorables2/totales
        proba3 = favorables3/totales
        proba4 = favorables4/totales
        proba5 = favorables5/totales
        proba6 = favorables6/totales
        
        # En 'renglon' va la información tal cual se va a escribir en el archivo, renglón por renglón.
        renglon = str(totales) + '\t' + str(tirada) + '\t' + str(proba1) + '\t' + str(proba2) + '\t' + str(proba3) + '\t' + str(proba4) + '\t' + str(proba5) + '\t' + str(proba6) + '\n'
        
        # escribimos el renglón en el archivo.
        archivo.write(renglon)
    
    # Siempre debemos cerrar las puertas que abrimos, así como los archivos que invocamos.
    archivo.close()
    
    # Esta función ya generó y escribió el archivo que queríamos, entonces no regresa nada.
    return
```

```python
#-----------------------------------------------------------------------------
#---------------------------------NUEVE---------------------------------------
# Vamos a escribir una función que aproxime el valor de \pi con el método de Montecarlo.
def problema9(num_de_iteraciones):
        
    # En 'casos_totales' guardaremos el número de puntos que hemos generado.
    casos_totales = 0
    
    # En 'casos_exitosos' guardaremos los puntos que están dentro del círculo unitario.
    casos_exitosos = 0
    
    # En este ciclo en sí generaremos los puntos y veremos si están dentro del círculo o no.
    for i in range(num_de_iteraciones):
        
        # Aquí está la pareja ordenada de puntos generados.
        x = rd.uniform(0, 1)
        y = rd.uniform(0, 1)
        
        # Con este 'if' calculamos si esta pareja está o no dentro del círculo unitario, en caso de estarlo contamos un punto exitoso más.
        if y <= np.sqrt(1 - x**2):
            casos_exitosos += 1
        
        # En cada corrida del ciclo contamos un punto generado más.
        casos_totales += 1
    
    # El área es el cociente de los puntos dentro del círculo y los puntos totales.
    area = casos_exitosos/casos_totales
    
    # Esta área es simplemente \pi/4 porque es solo un cuarto del círculo unitario, \pi es 4 aproximadamente cuatro veces el cociente que acabamos de encontrar.
    pi = 4*area
    
    # Esta función regresa el valor de pi
    return pi
```
